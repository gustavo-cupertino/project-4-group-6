```python
application_dummies = pd.get_dummies(application_df["HeartDisease"])
application_dummies.head()
```

```python
application_dummies["new_column"]=application_dummies["No"]+application_dummies["Yes"]
```

```python
application_dummies_cleaned_df=application_dummies.drop(columns=["No","Yes"])
```

```python
application_dummies_cleaned_df=application_dummies_cleaned_df.rename(columns={"new_column":"Outcome"})
application_dummies_cleaned_df
```

```python
application_dummies_2 = pd.get_dummies(application_df[["Smoking","AlcoholDrinking","Stroke","DiffWalking","Sex","Race","Diabetic","PhysicalActivity","GenHealth","Asthma","KidneyDisease","SkinCancer","AgeCategory"]])
application_dummies_2.head(20)
```

```python
merged_clean_df=pd.concat([application_dummies_cleaned_df,application_dummies_2],axis=1)
```

```python
merged_clean_df.head(10)
```

```python
application_copy_df=application_df.copy()
```

```python
merged_clean_df_2=pd.concat([application_copy_df,merged_clean_df],axis=1)
```

```python
merged_clean_df_2
```

```python
droped_merged_df=merged_clean_df_2.drop(columns=["HeartDisease","Smoking","AlcoholDrinking","Stroke","DiffWalking","Sex","AgeCategory","Race","Diabetic","PhysicalActivity","GenHealth","Asthma","KidneyDisease","SkinCancer"])
```

```python
droped_merged_df.head(20)
```

```python
droped_merged_df.info()
```

```python
y=droped_merged_df["Outcome"].values
```

```python
X=droped_merged_df.drop(columns=["Outcome"]).values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, random_state=78)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  8
hidden_nodes_layer2 = 5

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
fit_model = nn.fit(X_train_scaled,y_train,epochs=5)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
