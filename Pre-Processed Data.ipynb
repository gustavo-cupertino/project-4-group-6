```python
#  Import and read the charity_data.csv.
import pandas as pd 
data_df = pd.read_csv("Resources/heart_2020_cleaned.csv")
data_df.head(20)
```

```python
data_copy_df=data_df.copy()
```

```python
def outcome(results):
    if results == "No":
        return 0
    if results == "Yes":
        return 1
```

```python
data_copy_df["HeartDisease"]=data_copy_df["HeartDisease"].apply(outcome)
```

```python
data_copy_df.head(10)
```

```python
data_copy_df.describe()
```

```python
data_copy_dummies_df = pd.get_dummies(data_copy_df[["Smoking","AlcoholDrinking","Stroke","DiffWalking","Sex","Race","Diabetic","PhysicalActivity","GenHealth","Asthma","KidneyDisease","SkinCancer","AgeCategory"]])
data_copy_dummies_df.head(20)
```

```python
droped_data_df=data_copy_df.drop(columns=["Smoking","AlcoholDrinking","Stroke","DiffWalking","Sex","AgeCategory","Race","Diabetic","PhysicalActivity","GenHealth","Asthma","KidneyDisease","SkinCancer"])
```

```python
droped_data_df.head()
```

```python
merged_data_df=pd.concat([droped_data_df,data_copy_dummies_df],axis=1)
```

```python
merged_data_df.to_csv("Resources/Cleaned_Data.csv", index=False)
```
